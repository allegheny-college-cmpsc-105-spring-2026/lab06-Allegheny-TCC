```python
# data_survey.csv

# update the column names to match the names you have in your csv file for data_survey

# create a pandas dataframe
dfs = pd.read_csv(
    '/content/drive/MyDrive/data_survey.csv',
    sep=',',
    names = ['timestamp', 'age', 'favorite_color', 'top_musician'],
    skiprows=1
)

# Display the first few rows of the DataFrame
print(dfs.head())
```

```python
average_age_by_color = dfs.groupby('favorite_color')['age'].mean()
# print(dfs['average_age_by_color'])
print(average_age_by_color)
```

```python
# Analytics

# Do one or more explorations with dfa
# no interaction is required

# make a blank canvas
plt.figure()
# make a title starting with 'Figure 1:' Fill in a relevant title
plt.title('Figure 1: Event Count by Non-Set vs Organic Users')

# plot
plt.bar(dfa['session_manual'], dfa['event_count'])
plt.xlabel('Organic vs Non-Set')
plt.ylabel('Event Count')
plt.show()
# reveal everything
```
